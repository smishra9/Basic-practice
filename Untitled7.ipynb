```python
### Now there are some categorical variables which are useful for defining relation between values and other predictors
####Converting those variables into numerical data
length = {'2-year': 0,'4-year': 1}
duration = college.Length.map(length)
college['Length'] = duration
college
```

```python
###The same should be done with Expense
exp = {'Fees/Tuition':1,'Room/Board':0}
expenses= college.Expense.map(exp)
college['Expense'] = expenses
```

```python
college.head(3)
```

```python
###We want to see the relation between Type and VALUES
val = college.groupby('Type')[['Value']].mean()
sns.barplot(val.index,val.Value, x ='Tyoe of College',y ='Value/Fees')
```

###Its clearly visible that Private Fees is maximum and Public In state is minimum and we can conclude that there is relation between type and values in dataset

```python
###Now we also want to change Type of college to numerical data as seen above in the graph
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
x = preprocessing.OneHotEncoder()
x.fit(college[['Type']])
x.categories_
```

```python
one_hot = x.transform(college[['Type']]).toarray()
one_hot
```

```python
college[['Private','Public In-State','Public Out-of-State']] = one_hot
```

```python
college
```

```python
###Now we will perform multiple regression to train the model
input = ['Year','Length','Expense','Private','Public In-State','Public Out-of-State']
inputs,target = college[input],college['Value']
```

```python
model = LinearRegression().fit(inputs,target)
pred= model.predict(inputs)
```

```python
coef = model.coef_
print(coef)
inter = model.intercept_
print(inter)
```

```python
###We want to find the loss that is root mean square error
from numpy import sqrt
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(target,pred)
rmse = sqrt(mse)
print('The loss for our model is :', rmse)
```
