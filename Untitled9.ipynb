```python
adult_subset=adult[['age', 'gender', 'hours-per-week', 'income']]
adult_subset
```

```python
adult_subset.gender.unique()
```

```python
adults = {'Male':1,'Female':0}
ad = adult_subset.gender.map(adults)
adult_subset['gender'] = ad
```

```python
adult_subset
```

```python
adult_subset.income.unique()
```

```python
inc = {'<=50K':0,'>50K':1}
sdm = adult_subset.income.map(inc)
adult_subset['income'] = sdm
```

```python
adult_subset
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(adult_subset[['age', 'gender', 'hours-per-week']],adult_subset.income,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
x_test
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
