```python
### Now changing the salary to numeric
from sklearn import preprocessing
sdm = preprocessing.OneHotEncoder()
sdm.fit(data[['Department']])
sdm.categories_
```

```python
ot = sdm.transform(data[['Department']]).toarray()
```

```python
data[['IT', 'RandD', 'accounting', 'hr', 'management', 'marketing',
        'product_mng', 'sales', 'support', 'technical']] = ot
```

```python
data.head()
```

```python
data.salary
```

```python
### Now doing data partition
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data[['satisfaction_level','number_project','average_montly_hours','time_spend_company','IT',	'RandD','accounting','hr','management','marketing','product_mng','sales','support','technical']],data.salary,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
abc = model.predict(x_test)

abc
list1 = abc.tolist()
```

```python
model.score(x_test,y_test)
```
